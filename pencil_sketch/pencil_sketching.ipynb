```python
# import the librarires
import cv2
import sys
import matplotlib.pyplot as plt

```

```python
# read the image using cv package
image = cv2.imread("messi.jpg")
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# convert to inverse gray scale image
grayImageInv = 255 - grayImage
grayImageInv = cv2.GaussianBlur(grayImageInv, (21, 21), 0)
output = cv2.divide(grayImage, 255 - grayImageInv, scale=256.0)

origImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
pencil_im = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
```

```python
# Plotting the orignal and pencil sketched image
fig, axs = plt.subplots(1, 2)
fig.set_figheight(15)
fig.set_figwidth(15)

fig.subplots_adjust(top=0.95)
fig.tight_layout()
#fig.suptitle("pencil sketch from RGB image ")
axs[0].imshow(origImage)
axs[0].set_title('original image (RGB)')

axs[1].imshow(pencil_im)
axs[1].set_title('Pencil sketch image (RGB)')

```
